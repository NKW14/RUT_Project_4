```python
from pathlib import Path
import pandas as pd
```

```python
data = Path('Ethereum_10_17_2023-10_16_2024_historical_data_coinmarketcap.csv')
Eth_df = pd.read_csv(data, delimiter= ';')
Eth_df['timestamp'] = pd.to_datetime(Eth_df['timestamp'])
Eth_df['Date'] = Eth_df['timestamp'].dt.strftime('%m/%d/%Y')
Eth_df.head()
```

```python
Eth_df_clean = Eth_df[['Date','close', 'volume']]
Eth_df_clean = Eth_df_clean.rename(columns={'close': 'Close', 'volume': 'Volume'})
Eth_df_sorted = Eth_df_clean.sort_values(by='Date', ascending=True)
Eth_df_sorted.head()
```

```python
data = Path('Full_timeline_analysis\ETH-USD (2017-2024)_2.csv')
Eth_df_2 = pd.read_csv(data)
Eth_df_2['Date'] = pd.to_datetime(Eth_df_2['Date'])
Eth_df_2['Date'] = Eth_df_2['Date'].dt.strftime('%m/%d/%Y')
Eth_df_clean_old = Eth_df_2[['Date', 'Close','Volume']]
Eth_df_clean_old.head()
```

```python
final_2 = pd.concat([Eth_df_clean_old, Eth_df_sorted], axis=0)
e_final = final_2.sort_values(by='Date', ascending=True)
df_cleaned = e_final.drop_duplicates(subset=['Date'], keep='first')
df_cleaned.to_csv('total_Ethcoin.csv', index=False)
```
